```python
def retry_with_backoff(prompt, max_tokens, retries = 10, backoff_in_seconds = 1):
    x = 0
    while True:
        try:
            return client.chat.completions.create(
                  model="gpt-4",
                  messages=[
                    {"role": "user", "content": prompt},
                  ],
                  max_tokens = max_tokens,
                  temperature = 0,
                  frequency_penalty = 0,
                  presence_penalty =  0,
            )
        except:
            if x == retries:
                raise

            sleep = (backoff_in_seconds * 2 ** x +
                   random.uniform(0, 1))
            time.sleep(sleep)
            x += 1
```
